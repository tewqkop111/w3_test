```python
#Plot graph curve y=f(x)=x
import numpy as np
import matplotlib.pyplot as plt
width=0.001
# Create data
x=np.arange(0,1,width) # interval from 0 to 6 with unit 0.1
y=x**2
# plot graph
plt.plot(x,y)
plt.xlim(0,1)  # set the xlim to left, right
plt.ylim(0,1)
plt.show()
```

```python
objects =  y
y_pos = np.arange(len(objects))
plt.bar(y_pos, y, align='center', width=1,alpha=0.5)
plt.plot(y_pos, y)
plt.xticks(y_pos, objects)
```

We explore the definite integration of function $f(x)=x$:

$$
\begin{align*} 
I  &=\int_{0}^{1} x^2 dx \\
   &=\tfrac{1}{3} x^3 
\Big|_0^1=\tfrac{1}{3} 1^3 -0=\tfrac{1}{3}
\end{align*}
$$

The result I=0.33.
