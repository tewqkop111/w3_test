```python
import numpy as np
import matplotlib.pyplot as plt

# 目標函數:y=5*x**2+20*x+10
def func(x): return 5*x**2+20*x+10

# 目標函數一階導數:dy/dx=10*x+20
def dfunc(x): return 10*x+20

def GD(x_start, df, epochs, lr):
    xs = np.zeros(epochs+1)    
    x = x_start    
    xs[0] = x    
    for i in range(epochs):         
        dx = df(x)        
        # v表示x要改變的幅度        
        v =  - dx * lr        
        x += v        
        xs[i+1] = x    
    return xs

# Main
# 起始權重
x_start = 4
# 執行週期數
epochs = 30
# 學習率   
lr = 0.14
# 梯度下降法 
x = GD(x_start, dfunc, epochs, lr) 
print (x)

color = 'r' 
from numpy import arange
t = arange(-10, 6, 0.01)
plt.plot(t, func(t), c='b')
plt.plot(x, func(x), c=color, label='lr={}'.format(lr))    
plt.scatter(x, func(x), c=color, )    
plt.legend()

plt.show()
```
